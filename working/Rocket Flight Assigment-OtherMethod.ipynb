```python
import numpy as np
```

Variables:

```python
m_shell=50
g=9.81
ro=1.091
A=np.pi*0.5**2
Cd=0.15
m_fuel0=100
Ve=325
```

```python
def mp_dot(time):
    if time<=5:
        value=20
    else:
        value=0
    return value
```

```python
def mp(time):
    if time<=5:
        value=100-20.*time
    else:
        value=0.
    return value
```

Answering remaining fuel question:

```python
q1=mp(3.2)
print(q1)
```

Initial values:

```python
v0=0
h0=0
time0=0
```

Calculating

```python
time=time0
dt=0.000001
stop_time=16
timesteps=int((stop_time-time)/dt)

v=np.zeros([timesteps+2])
h=np.zeros([timesteps+2])
n=0
```

```python
while True:
    v[n+1]=v[n]+dt*(-g+mp_dot(time)/(m_shell+mp(time))*Ve-0.5*(ro/(m_shell+mp(time)))*v[n]*abs(v[n])*A*Cd)
    h[n+1]=h[n]+dt*v[n]
    
    time+=dt
    n+=1
    
    if v[n]<=0.: break
    
```

Answering maximum velocity question:

```python
Vmax=max(v)
print('Vmax = ',Vmax)
```

Time it occurs:

```python
for i,j in enumerate(v):
    if v[i]==Vmax:
        Vmax_time=i*dt+time0
        Vmax_index=i
        print(Vmax_index)
```

```python
print('Time it occurs =',Vmax_time)
```

Altitude:

```python
print(' It occurs in %.3f meters' %h[Vmax_index])
```
